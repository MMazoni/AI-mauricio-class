```python
import matplotlib
matplotlib.use('Agg') 
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
class combinacaoLinear(object):
    def __init__ (self, f1, f2, f3, f, pesos, taxaAprendizado):
        self.f1 = f1
        self.f2 = f2
        self.f3 = f3
        self.f = []
        self.pesos = pesos
        self.taxaAprendizado = taxaAprendizado
        self.t = [x for x in range(0,300)]
        
    def entrada(self, w):
        for ix in range(0,300):
            self.t[ix] = w[0] + w[1]*f1[ix] + w[2]*f2[ix] + w[3]*f3[ix]
            try:
                w[0] += self.taxaAprendizado*math.pow( (self.f[ix]-self.t[ix]), 2.0)*0.5
                w[1] += self.taxaAprendizado*math.pow( (self.f[ix]-self.t[ix]), 2.0)*0.5*self.f1[ix]
                w[2] += self.taxaAprendizado*math.pow( (self.f[ix]-self.t[ix]), 2.0)*0.5*self.f2[ix]
                w[3] += self.taxaAprendizado*math.pow( (self.f[ix]-self.t[ix]), 2.0)*0.5*self.f3[ix]
            except Exception as e:
                pass
        return w
        
    def iteracao(self):
        for ix in range(0, 300):
            # f1[ix]=(math.sin(math.radians(ix)))
            # f2[ix]=(math.cos(math.radians(ix)))
            # f3[ix]=
            try:
                self.f.append(-math.pi + 0.565*self.f1[ix] + 2.657*self.f2[ix] + 0.674*self.f3[ix]) # exemplo da aula
            except Exception as e:
                pass
            # self.f.append(self.pesos[0] + self.pesos[1]*f1[ix] + self.pesos[2]*f2[ix] + self.pesos[3]*f3[ix])
        for i in range(0, 32):
            pesoAntigo = self.pesos
            self.pesos = self.entrada(pesoAntigo)
            
    
    def testa_af(self, arrPesos):
        arrF = [x for x in range(0, 300)]
        arrG = [x for x in range(0, 600)]
        for i in range(0, 300):
            arrF[i] = arrPesos[0] + arrPesos[1]*self.f1[i] + arrPesos[2]*self.f2[i] + arrPesos[3]*self.f3[i]
        arrG[0:300] = arrF
        k = 0
        for j in range(300, 600):
            arrG[j] = arrF[k]
            k += 1
        return arrG
```

```python
## a)
#   f4[i]= (Math.sin(i*Math.PI/180))*0.8
#   f5[i]= (Math.cos(i*Math.PI/180))*0.9
#   f6[i]= (i*Math.PI/180)*1.1
#   f7[i] = ww[0][0] + ww[1][0]*f4[i] + ww[2][0]*f5[i] + ww[3][0]*f6[i]
plt.clf()
f1 = []
f2 = []
f3 = []
for ix in range(0, 300):
    f1.append(math.sin(math.radians(ix))*0.8)
    f2.append(math.cos(math.radians(ix))*0.9)
    f3.append(math.radians(ix)*1.1)

adaline = combinacaoLinear(f1, f2, f3, [], [-2.4013, 0.393, 1.902, 0.429], 0.000027 )
adaline.iteracao()
pesos = adaline.testa_af(adaline.pesos)

plt.plot([x for x in range(0,300)], adaline.f, 'ro')
plt.plot([x for x in range(0,300)], adaline.t, 'bo')
plt.savefig('letraA.png')
```
